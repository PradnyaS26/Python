Q.2) Arithmetic Operations on float

Take two different float values. Store them in two different variables.

Do below operations on them:-
* Find sum of both numbers
* Find difference between them
* Find the product of both numbers.
* Find value after dividing first num with second number
* Find the remainder after dividing first number with second number
* Find the quotient after dividing first number with second number
* Find the result of the first num to the power of the second number.

```python
float2 = 3.5
float3 = 5.6
sum = float2 + float3
difference = float2 - float3
product = float2 * float3
floorDivision = float2 // float3
remainder = float2 % float3
quotient = float2 / float3
exponential = float2 ** float3
print("Sum : {}".format(sum))
print("Difference : {}".format(difference))
print("Product : {}".format(product))
print("Floor Division : {}".format(floorDivision))
print("Remainder : {}".format(remainder))
print("Quotient : {}".format(quotient))
print("Exponential : {}".format(exponential))

# Output :

# Sum : 9.1
# Difference : -2.0999999999999996
# Product : 19.599999999999998
# Floor Division : 0.0
# Remainder : 3.5
# Quotient : 0.625
# Exponential : 1113.7328982650954
```

Q.3) Comparison Operators on float

Take two different float values. Store them in two different variables.

Do below operations on them:-

Compare the two numbers with below operator:-

* Greater than, '>'
* Smaller than, '<'
* Greater than or equal to, '>='
* Less than or equal to, '<='

Observe their output(return type should be boolean)

```python
float4 = 6.6
float5 = 9.8
print("float4 > float5: {}".format(float4 > float5))
print("float4 < float5: {}".format(float4 < float5))
print("float4 >= float5 : {}".format(float4 >= float5))
print("float4 <= float5 : {}".format(float4 <= float5))

# Output :

# float4 > float5: False
# float4 < float5: True
# float4 >= float5 : False
# float4 <= float5 : True
```

Q.8) Membership operation

in, not in are two membership operators and it returns boolean value
* print('2.7' in 'Python2.7.8') #True
* print(10.20 in [10,10.20,10+20j,'Python']) #True
* print(10.20 in (10,10.20,10+20j,'Python')) # True
* print(20.30 in {1,20.30,30+40j}) # True
* print(2.3 in {1:100, 2.3:200, 30+40j:300}) # True
* print(10 in range(20)) # True

```python
print('2.7' in 'Python2.7.8')
print(10.20 in [10,10.20,10+20j,'Python'])
print(10.20 in (10,10.20,10+20j,'Python'))
print(20.30 in {1,20.30,30+40j})
print(2.3 in {1:100, 2.3:200, 30+40j:300})
print(10 in range(20))

# Output :

# True
# True
# True
# True
# True
# True

# Membership operator return True if given value exists in the given sequence
```
